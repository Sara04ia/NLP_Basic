## Tokenization
The task of segmenting text into relevant words is called tokenization.

Tokenization is the process of breaking down the given text in natural language processing into the smallest unit in a sentence called a token. 

0_EKgminT7W-0R4Iae.png

```python
import nltk #natural language toolkit
from nltk.tokenize import word_tokenize ,sent_tokenize #to break the text into sentences/words
```

#### Text is tokenize into a list of sentences 

```python
text_Eng = "Hello, my name is Sara, I am a student of data science and analysis at the University of Jeddah"
sent_Eng = sent_tokenize(text_Eng)
print(sent_Eng) 
```

```python
text_arb ="مرحبًا اسمي سارة ، انا طالبة في جامعة جدة بتخصص علوم البيانات وتحليلها"
sent_arb = sent_tokenize(text_arb)
print(sent_arb) 
```

#### Text is tokenize into a list of Words in every sentences as a single word 

```python
sentences_word = [word_tokenize(sent) for sent in sent_Eng]
print(sentences_word)
```

```python
sentences_word_ar = [word_tokenize(sent) for sent in sent_arb]
print(sentences_word_ar)
```

 #### Text is tokenize into a list of Words 

```python
words_En = word_tokenize(text_Eng)
print(words_En)
```

```python
words_Ar = word_tokenize(text_arb)
print(words_Ar) 
```

## Stop Words
When we deal with text problems in Natural Language Processing, Stop words removal process is one of the important steps to have a better input for any models. Stop words mean that it is a very common word in a language

cfeb9180-6dde-11e9-85b6-e79357766310.jfif

```python
#linguistic resources 
from nltk.corpus import stopwords 

#treated as separate tokens by NLTK when tokenization happens
from string import punctuation

#create a custom set of stopwords
customstopword = set(stopwords.words('english')+list(punctuation))
```

```python
EnWOstopword = [word for word in word_tokenize(text_Eng) if word not in customstopword ]
print(EnWOstopword) # all the stop word are removed in this step 
```

## Stemming
Stemming is an operation which uses stemming algorithms and reduces the words into it’s base or root word.


main-qimg-187b045c480fa7c0b16869daa0661b5a.webp

```python
from nltk.stem.lancaster import LancasterStemmer #lancaster stemmer is algorithm 
st = LancasterStemmer()
```

```python
text_en2 = "Hello my name is sara, my closing friends are Ebtesam, Hend, and Jana they are the closest friends in my life"
stemmedWords = [st.stem(word)for word in word_tokenize(text_en2)]
print(stemmedWords)
```

## Part of speech Tagging 
Part-of-speech (POS) tagging is a popular Natural Language Processing process which refers to categorizing words in a text (corpus) in correspondence with a particular part of speech, depending on the definition of the word and its context.

pos-title.jpg

```python
nltk.pos_tag(word_tokenize(text_en2))
```

take each one of the word and tag them to the relevant part of speech noun , verb , adverb and so on
